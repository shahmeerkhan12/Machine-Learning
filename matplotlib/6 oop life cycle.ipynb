```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../heart.csv')
dataset.head()
```

**select all patients above 40 years of age**

```python
above_40 = dataset[dataset['age']>40]
above_40
```

**how to draw to plots inside a single figure**

```python
fig, (plot1, plot2) = plt.subplots(nrows = 1, ncols = 2, figsize=(10,3))

# now create the first plot
scatter_plot = plot1.scatter(x=above_40['age'],y=above_40['chol'],c= above_40['target'])
plot1.set(title="Heart Disease Probability", xlabel  = 'age', ylabel = 'cholestrol')
plot1.legend(*scatter_plot.legend_elements(),title = 'survived')

# creating the second plot
scatter_plot = plot2.scatter(x=above_40['age'],y=above_40['ca'],c= above_40['target'])
plot2.set(title = "Heart Disease Probability", xlabel  = 'age', ylabel = 'block vessels')
plot2.legend(*scatter_plot.legend_elements(), title = 'survived')
```

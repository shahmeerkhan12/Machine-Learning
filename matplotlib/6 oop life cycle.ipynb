```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../heart.csv')
dataset.head()
```

**select all patients above 40 years of age**

```python
above_40 = dataset[dataset['age']>40]
above_40
```

**how to draw to plots inside a single figure**

```python
fig, (plot1, plot2) = plt.subplots(nrows = 2, ncols = 1, figsize=(10,10), sharex = True)

# now create the first plot
# cmap is used for coloring markers in the plot
scatter_plot = plot1.scatter(x=above_40['age'],y=above_40['chol'],c= above_40['target'],cmap = 'copper')
plot1.set(title="Cholestrol vs Age", ylabel = 'cholestrol' )
plot1.legend(*scatter_plot.legend_elements(),title = 'survived')

# creating the second plot
scatter_plot = plot2.scatter(x=above_40['age'],y=above_40['ca'],c= above_40['target'],cmap = 'RdGy')
plot2.set(title = "Block Blood Vessels vs Age",xlabel='age', ylabel = 'block vessels')
plot2.legend(*scatter_plot.legend_elements(), title = 'survived')

# adding some customization

# 01 mean line
plot1.axhline(y=above_40['chol'].mean(), linestyle = "--");
plot2.axhline(y=above_40['ca'].mean(), linestyle = "--");

# plot styling
plt.style.use('seaborn-v0_8-darkgrid');
        
# seting the fig title
fig.suptitle("Heart Disease Patients Dataset Insights",fontsize = 32, fontweight = 'bold')

# saving the fig
fig.savefig("Patients Data.png")
```

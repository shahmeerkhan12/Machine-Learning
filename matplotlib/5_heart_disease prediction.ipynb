```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart_dataset = pd.read_csv("../heart.csv")
```

```python
heart_dataset.head()
```

```python
above_40 = heart_dataset[heart_dataset['age']>40]
above_40
```

**ploting the scatter plot for people of age above 40 years**

```python
above_40.plot(kind='scatter',x='chol',y='age',c='target');
```

# the object oriented ploting method life cycle

```python
# plot the above 40 with
fig, pred_prob = plt.subplots(figsize=(6,4))
scatter_plot = pred_prob.scatter(x=above_40['age'],y=above_40['chol'],c=above_40['target'])
pred_prob.set(title="Heart Disease Prediction based on Age and Chol",xlabel="age",ylabel="cholestrol");
pred_prob.legend(*scatter_plot.legend_elements(),title='target')
scatter_plot;
```

```python
import matplotlib.pyplot as plt
```

# subplots and figures

```python
temp = [34,39,34,38,29,33,38]
humidity = [.34,.1,.2,1.0,.06,.9,.4]
age = {
    "ali":20,
    'ahmad':28,
    'usman':13,
    'uzair':23
}
day = ['mon','tue','wed','thu','fri','sat','sun']
```

# line plot, scatter plot, histogram, bar chart

```python
import numpy as np
fig, ((subplot1,subplot2),(subplot3,subplot4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
subplot1.plot(day,temp)
subplot2.scatter(day,humidity)
subplot3.hist(np.random.randn(1000)) # histogram shows gausian distribution of a random dataset
subplot4.bar(age.keys(),age.values()) # takes key value pairs
plt.show()
```

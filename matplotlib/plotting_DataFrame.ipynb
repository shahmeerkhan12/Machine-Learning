```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../phones.csv')
```

```python
dataset
```

# turning the numerical text columns to real numbers

```python
dataset['Memory'], dataset['Storage'] = dataset['Memory'].str.replace('GB','',regex=False),dataset['Storage'].str.replace('GB','',regex=False)
```

```python
dataset['Memory'],dataset['Storage'] = dataset['Memory'].astype(int),dataset['Storage'].astype(int)
```

```python
fig, (plot1,plot2) = plt.subplots(1,2,figsize=(10,5))
plot1.bar(dataset['Manufacturer'],dataset['Price (USD)'])
plot1.set_title('Phone Manufacturer Vs Prices(USD)')
plot1.set_xlabel('Manufacturer')
plot1.tick_params(axis='x', rotation=60)
    
plot2.bar(dataset['Phone Name'],dataset['Storage'])
plot2.set_title('Phone vs Storage')
plot2.set_xlabel("phone")
plot2.set_ylabel=("storage")
plot2.tick_params(axis='x',rotation=45)
plt.tight_layout()
plt.show()
```

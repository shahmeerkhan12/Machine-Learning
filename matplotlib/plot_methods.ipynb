```python
import matplotlib.pyplot as plt
```

# the different methods of PLOTING

**method 1**

```python
# let we have a common data for all the plots: we will plot the same data using multiple methods
x= [19,25,38,42,53,67,77,89,98]
y= [19,15,18,12,13,17,13,14,15]
```

```python
home = plt.figure()
room = home.add_subplot()
room.plot(x,y)
plt.show()
```

**method 2**

```python
home = plt.figure()
room = home.add_axes([1,1,1,1])
room.plot(x,y)
plt.show()
```

**method 3**

```python
# short and quik method
home, room = plt.subplots()
room.plot(x,y)
plt.show()
```

# plot labels and title

```python
home, room = plt.subplots()
room.plot(x,y)
room.set(title="Days vs Patients",
        ylabel='days',
        xlabel='patients')
plt.show()
```

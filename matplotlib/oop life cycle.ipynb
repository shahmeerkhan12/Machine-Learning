# Object Oriented PLotting life Cycle

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart_dataset = pd.read_csv("../heart.csv")
```

```python
over_40 = heart_dataset[heart_dataset['age']>40]
```

```python
over_40
```

**step-1: define the figure and plotting area/ graph area**

```python
fig, graph = plt.subplots(figsize=(6,4))
```

**step-2: plot the dataset/data**

```python
ageVsChol_scatter = graph.scatter(x=over_40['age'],y=over_40['chol'],c=over_40['target'])
```

**step-3: set labels and title of the graph**

```python
graph.set(title="Heart Disease Prediction (age vs cholestrol)",xlabel='age',ylabel='cholestrol')
```

**step-4: set legends**

```python
graph.legend(*ageVsChol_scatter.legend_elements(),title='target');
```

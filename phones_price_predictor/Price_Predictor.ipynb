```python
# Simple droping nan values, but now u can specify column name 
# ( does not mean to drop a column but instead the nan values with in it and the row in which it is present)
phone_ds.dropna(subset = ['Price (USD)'],inplace= True)
# CHECK IF IT WORKS
phone_ds.isna().sum()
```

```python
# but first let me split the data
X = phone_ds.drop('Price (USD)', axis = 1)
Y = phone_ds['Price (USD)']
```

```python
# the other solution could be filling missing values, now sklearn will help me do that
# imputation(filling missing values)

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# i will filling categorical data with "missing" and numerical with the mean values but here it is a bit different
cat_feature = SimpleImputer(strategy="constant", fill_value="missing")
simCard_feature = SimpleImputer(strategy="constant", fill_value="4")
num_feature = SimpleImputer(strategy="mean")

# filling the columns
cat_col = ["Manufacturer","OS"] # if i have more than one column
num_col = ["Storage (GB)","RAM (GB)"]
imputer =ColumnTransformer([("cat_feature",cat_feature,cat_col),("simCard_feature",simCard_feature,"Internet Network"),("num_feature",num_feature,num_col)])

filled_X = imputer.fit_transform(X)
```

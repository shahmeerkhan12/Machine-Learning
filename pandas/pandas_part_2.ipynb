```python
import pandas as pd
```

```python
phones = pd.read_csv("mobilePhones.txt")
```

```python
phones.iloc[:3]
```

```python
#let me compare two columns/features data: df.crosstab(col1,col2)
pd.crosstab(phones['Memory'],phones['Storage'])
```

```python
phones['Memory']=phones['Memory'].str.replace('GB','',regex = False).astype(int)
```

```python
phones['Storage']=phones['Storage'].str.replace('GB','',regex = False).astype(int)
```

```python
phones
```

```python
# let me compare different mobiles phones based on their memory, storage and price etc 
grouped_data = phones.groupby('Manufacturer').agg({
    'Storage': 'mean',
    'Memory': 'mean',
    'Price (USD)': 'mean'
})
```

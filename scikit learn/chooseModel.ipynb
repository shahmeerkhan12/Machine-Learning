# SKLEARN PROJECT A to Z, let us talk some fundamentals

# Load a built-in dataset from sklearn

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
```

## Turn the ds into a dataframe for model training

```python
iris_df = pd.DataFrame(iris["data"],columns=iris["feature_names"])
iris_df["target"]= pd.Series(iris["target"])
iris_df.head()
```

## Split the data

```python
from sklearn.model_selection import train_test_split
x = iris_df.drop("target",axis=1)
y = iris_df["target"]
np.shape(x),np.shape(y)
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(x,y,test_size = 0.2)
```

## Choose and train the model

```python
from sklearn.linear_model import Ridge

# create an object of the Rige class
ridge = Ridge()

# TRAIN ON DATA
ridge.fit(X_TRAIN,Y_TRAIN)

# ACCURACY TRAIN VS TEST
ridge.score(X_TRAIN,Y_TRAIN), ridge.score(X_TEST,Y_TEST)
```

### TRAINING ACCURACY: 0.9188
### VALIDATION ACCURACY: 0.9634

# SUMMARY

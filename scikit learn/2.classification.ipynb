# choosing best model : CLASSIFICATION PROBLEM

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# import the heart-disease datasets, b/c it is a classification problem
heart_ds = pd.read_csv('./../heart.csv')
```

```python
heart_ds.head()
```

## now check the sklearn model selection map, and consider these details:
### 1. do u want to predict a category: yes
### 2. nature of the problem : classfication
### 3. dataset examples i.e. 303 in this case
### 4. is data labeled: yes in this case

```python
# from above details, linear SVC model would be better compared to other
from sklearn.svm import LinearSVC

# train test splitting
x = heart_ds.drop("target",axis=1)
y = heart_ds["target"]

# split the data into train and  test 
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(x,y,test_size = 0.3)

# fit to model
# create an object of the Rige class
Linear_svc_model = LinearSVC()

# TRAIN ON DATA
Linear_svc_model.fit(X_TRAIN,Y_TRAIN)

# ACCURACY TRAIN VS TEST
Linear_svc_model.score(X_TRAIN,Y_TRAIN), Linear_svc_model.score(X_TEST,Y_TEST)
```

## what if you are not satisfied with the accuracy of the current model(above),
# then move to another one called Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# np.random.seed(1)

# train test splitting
x = heart_ds.drop("target",axis=1)
y = heart_ds["target"]

# split the data into train and  test 
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(x,y,test_size = 0.2)

# fit to model
# create an object of the Rige class
RFClassifier = RandomForestClassifier()

# TRAIN ON DATA
RFClassifier.fit(X_TRAIN,Y_TRAIN)

# ACCURACY TRAIN VS TEST
RFClassifier.score(X_TRAIN,Y_TRAIN), RFClassifier.score(X_TEST,Y_TEST)
```

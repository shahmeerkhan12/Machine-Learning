# Handling MIssing Values

```python
import pandas as pd
import numpy as np
```

```python
# import the dataset
dataset = pd.read_csv('missingData.txt')
dataset
```

```python
# in case of filling a missing value in a column having string objects as the datatype
dataset["Storage"].fillna(12,inplace=True)
```

## Step 2
## If there is any missing value like nan(instead of a valid number), then try to fill it appropriate values

```python
# lets fill it

# in case of filling a missing value in a column having string objects as the datatype
dataset["sex"].fillna("MissingVaue",inplace=True)

# in case of filling a missing value in a column having "int" as the datatype
dataset["chol"].fillna(dataset["chol"].mean,inplace=True)
```

## Step 3
## Be careful while working with Target or the column to be Predicted, because we don't want to fill the predicted/target missing values, so simply drop it

```python
dataset.dropna(inplace=True)
```

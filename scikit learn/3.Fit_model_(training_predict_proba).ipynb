# fit model: model.fit => model training and model predictions

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
import numpy as np
iris_df = pd.DataFrame(iris['data'], columns = iris['feature_names'] )
iris_df['target']= pd.Series(iris['target'])
iris_df.head(-15)
```

```python
from sklearn.model_selection import train_test_split
X = iris_df.drop('target',axis = 1)
Y = iris_df['target']
```

```python
x_train,x_test, y_train , y_test = train_test_split(X,Y, test_size = 0.2)
```

## Remember this is a Classification Problem

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train);
```

```python
rfc.predict(x_test)[:5]
```

```python
rfc.predict_proba(x_test)[:5]
# remember the output shows the probability of each of the value, 0, 1, 2. if there exists a 1 in the first col and 0's in other it means the result is 0 (a category)
```
